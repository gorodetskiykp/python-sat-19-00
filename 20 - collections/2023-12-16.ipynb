```python
from random import shuffle


class Deck:
    def __init__(self):
        self.cards = [1, 2, 3, 4, 5]
        self.shuffle_the_deck()

    def shuffle_the_deck(self):
        shuffle(self.cards)

    def give_cards(self, needs):
        give_out = self.cards[:needs]
        self.cards[:needs] = []
        return give_out
```

```python
from collections import ChainMap

n1 = {n: n ** 2 for n in range(10) if n % 2 == 0}
n2 = {n: n ** 3 for n in range(10) if n % 2 == 1}

n1[3] = "HELLO"

nn = ChainMap(n1, n2)

nn[30] = "WORLD"
nn[30]
n1
```

```python
from collections import deque

dq = deque([1, 2, 3, 4, 5])
dq.append(6)
print(dq)
print()
dq.appendleft(0)  # lst.insert(0, 0)
print(dq)
print()
print(dq.pop())
print(dq)
print()
print(dq.popleft())  # lst.pop(0)
print(dq)
print()
dq.reverse()
print(dq)
print()
dq.reverse()
print(dq)
dq.rotate(1)
print(dq)
print()
```

```python
from collections import deque

dq = deque([1, 2, 3, 4, 5], maxlen=5)

dq.append(6)
print(dq)
print()

dq.appendleft(1)
print(dq)
print()
```

```python
from collections import namedtuple

t_user = ('Tom', 41)
print(t_user[1])

User = namedtuple('User', ['name', 'age'])
nt_user = User(name='Tom', age=41)
print(nt_user.age)
# nt_user.age = 42  # ERROR

args = {
    'name': 'Tom',
    'age': 41
}

nt_user = User(**args)  # (name='Tom', age=41)
print(nt_user.age)
```

```python
class User:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age
        self.age = int(self.age)
        

args = {
    'name': 'Tom',
    'age': '41'
}

cls_user = User(**args)  # (name='Tom', age=41)
print(cls_user.age + 1)
```

```python
from dataclasses import dataclass


@dataclass
class User:
    name: str
    age: int
        
    def __post_init__(self):
        self.age = int(self.age)
        
        
args = {
    'name': 'Tom',
    'age': '41'
}

cls_user = User(**args)  # (name='Tom', age=41)
print(cls_user.name, cls_user.age + 1)
```

```python
from pydantic import BaseModel, conint
from typing import Union, Optional


class User(BaseModel):
    name: Union[str, None]
    age: conint(gt=0)
    children: Optional[list[User]]


class Empl(User):
    proff: str

    
args = {
    'name': 'Tom',
    'age': '20'
}

pd_user = User(**args)  # (name='Tom', age=41)
print(pd_user.name, pd_user.age + 1)

empl = Empl(**args, proff='dev')
print(empl)
```
